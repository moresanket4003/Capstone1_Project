```python
male_weights = np.array(male_weights)
female_weights = np.array(female_weights)
```

```python
male_mean = np.mean(male_weights)
male_median = np.median(male_weights)
male_variance = np.var(male_weights)
male_std = np.std(male_weights)
male_iqr = np.percentile(male_weights, 75) - np.percentile(male_weights, 25)
male_skewness = skew(male_weights)
male_kurtosis = kurtosis(male_weights)
```

```python
female_mean = np.mean(female_weights)
female_median = np.median(female_weights)
female_variance = np.var(female_weights)
female_std = np.std(female_weights)
female_iqr = np.percentile(female_weights, 75) - np.percentile(female_weights, 25)
female_skewness = skew(female_weights)
female_kurtosis = kurtosis(female_weights)
```

```python
print("Male Weights:")
print(f"  Mean: {male_mean}")
print(f"  Median: {male_median}")
print(f"  Variance: {male_variance}")
print(f"  Standard Deviation: {male_std}")
print(f"  Interquartile Range (IQR): {male_iqr}")
print(f"  Skewness: {male_skewness}")
print(f"  Kurtosis: {male_kurtosis}\n")

print("Female Weights:")
print(f"  Mean: {female_mean}")
print(f"  Median: {female_median}")
print(f"  Variance: {female_variance}")
print(f"  Standard Deviation: {female_std}")
print(f"  Interquartile Range (IQR): {female_iqr}")
print(f"  Skewness: {female_skewness}")
print(f"  Kurtosis: {female_kurtosis}")
```

- Males have higher mean and median weights than females, indicating that on average, males weigh more than females in this dataset.

- Both males and females have similar variances and standard deviations, with females showing a slightly higher spread. The IQR is nearly the same for both genders.

- Both distributions are positively skewed, with females showing a higher skewness, indicating more extreme high weights relative to the mean.

- The kurtosis values suggest that both distributions have heavier tails than a normal distribution, with females having a slightly higher kurtosis.

- Hip Circumference (BMXHIP) and Waist Circumference (BMXWAIST) are strongly correlated with each other and with BMI, indicating a strong relationship among these body measurements.

## In your own words, list some advantages and disadvantages of BMI, waist-to-height ratio, and waist-to-hip ratio.

### BMI (Body Mass Index)

### What it measures: BMI gives a general idea of overall body fatness based on weight and height.

Advantages:

Easy to calculate and understand.
Widely used in health assessments.
Linked to health risks like heart disease and diabetes.
Disadvantages:

Doesn't distinguish between fat mass and muscle mass.
Doesn't account for where fat is located on the body.
Waist-to-Height Ratio (WHtR)

### What it measures: WHtR focuses on central obesity by comparing waist size to height.

Advantages:

Simple to interpret; a ratio below 0.5 is usually seen as healthier.
Works across different ethnicities and heights.
Strongly correlates with health risks like diabetes.
Disadvantages:

Exact cutoffs can vary slightly.
Like BMI, doesn't differentiate muscle from fat.
Doesn't consider hip size, which can affect health risks.
Waist-to-Hip Ratio (WHR)

### What it measures: WHR looks at fat distribution by comparing waist size to hip size.

Advantages:

Highlights fat distribution, especially dangerous belly fat.
Has specific guidelines based on sex for health risks.
Predicts heart disease and metabolic issues well.
Disadvantages:

Requires measuring both waist and hip.
Optimal ratios may differ among ethnic groups.
Less useful for certain groups like pregnant women.

## Print out the standardised body measurements for the 5 persons with the lowest BMI and the 5 persons with the 5 highest BMI (e.g., call print for a subset of zfemale comprised of 10 chosen rows as determined by a call to numpy.argsort). Interpret the results.

```python
lowest_bmi_indices = np.argsort(zfemale['BMI'])[:5]
highest_bmi_indices = np.argsort(zfemale['BMI'])[-5:]

print("Lowest Standardized BMI:-")
print(zfemale['BMI'].iloc[lowest_bmi_indices])

print("Highest Standardized BMI")
print(zfemale['BMI'].iloc[highest_bmi_indices])

# Combine these indices
selected_indices = np.concatenate((lowest_bmi_indices, highest_bmi_indices))

selected_measurements = zfemale.iloc[selected_indices]

print(selected_measurements)
```

### Magnitude of Deviation:

- The highest standardized BMIs show a much greater deviation from the mean compared to the lowest standardized BMIs. This indicates that the dataset includes individuals with BMI values that are significantly higher than the mean, suggesting the presence of extreme obesity cases.

### Health Implications:

- Low BMI Group: The individuals in this group are likely underweight. Being underweight can lead to health issues such as weakened immune systems, malnutrition, osteoporosis, and potential reproductive issues.

- High BMI Group: The individuals in this group are likely severely overweight or obese. This condition is associated with a higher risk of chronic diseases such as type 2 diabetes, hypertension, cardiovascular diseases, and certain cancers.

### Distribution Insights:

- The extreme deviation in the highest BMI values compared to the lowest indicates a potential skewness in the dataset towards higher BMI values. This might suggest that obesity is a more pronounced issue within this population compared to underweight cases.

### Comparative Analysis:

- The lowest standardized BMI values, while significantly below the mean, do not show as extreme deviations as the highest standardized BMI values. This suggests that while underweight individuals are present, the prevalence and severity of high BMI values are more pronounced.
